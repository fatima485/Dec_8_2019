```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df2
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a','a','b'],
                    'data1': range(7)})
df1
```

```python
pd.merge(df1, df2)
```

```python
df2['key1']=['a','d','c']
pd.merge(df1,df2)
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
df3
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
df3
```

```python
merged_result = pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
merged_result
```

```python
merged_result = pd.merge(df3, df4, left_on='lkey', right_on='rkey', how = 'outer')
```

```python
merged_result
```

```python
df3.join(df3, on=['lkey','rkey'])
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
right2
```

```python
left2.join(right2, how = 'outer')
```

```python
left2.join(right2, how = 'inner')
```

```python
left2.join(right2, how = 'right')
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df1
```

```python
df2 = pd.DataFrame(5 + np.arange(4).reshape(2,2),index=['a','c'],
                   columns=['one','two'])
df2
```

```python
pd.concat([df1,df2])
```
